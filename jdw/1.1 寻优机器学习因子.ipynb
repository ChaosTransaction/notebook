```python
import os
os.environ['DB_URL'] = 'mysql+mysqlconnector://ultron:123456@127.0.0.1/ultron'
```

```python
from jdw import EntropyAPI
```

```python
begin_date = '2021-06-10'
end_date = '2021-08-10'
```

```python
universe = 'zz500'
industry_name = 'sw'
industry_level = 1
thresh = 0.98 ## 降低因子之间相关性
```

```python
factor_columns = ['aiEtopZ180', 'aiDaPE60','aiDaNp60','hkHoldRatioB','DAVOL5','DHILO']
```

```python
stock_engine = EntropyAPI.StockCatalyst(thresh=thresh,
                                        universe=universe,
                                        factor_columns=factor_columns,
                                        industry_name=industry_name,
                                        industry_level=industry_level, yield_name='returns')
```

```python
stock_engine.create_params(generations=5)
```

```python
stock_engine.run(begin_date=begin_date, end_date=end_date, codes=None)
```
