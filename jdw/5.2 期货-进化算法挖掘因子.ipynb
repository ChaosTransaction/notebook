```python
from jdw import EntropyAPI
```

#### 设置参数

```python
#### 基础因子
factor_columns = [
    'Alpha191_20_10', 'Alpha191_66_6', 'Alpha191_41_10_10', 'Alpha101_53_9',
    'Alpha101_11_3'
]

#### 运算符
operators_sets = [
    'AVG', 'DIFF', 'LOG', 'SQRT', 'ABS', 'ACOSH', 'ASINH', 'CEIL', 'FLOOR',
    'ROUND', 'RETURNSimple', 'RETURNLog', 'ADDED', 'SUBBED', 'MUL', 'DIV',
    'MINIMUM', 'MAXIMUM', 'CSMean', 'CSRes', 'EMA', 'MA', 'MADecay', 'MMAX',
    'MMIN', 'MRANK', 'MQUANTILE', 'MSUM', 'MVARIANCE', 'MSTD', 'MNPOSITIVE',
    'MAPOSITIVE', 'RSI', 'MARETURNLog', 'DELTA', 'SHIFT', 'MCORR'
]

#### 配置参数
configure = {'evaluate': 'both_evaluate', 'method': 'fitness', 'generations':3} # 评估函数

offset = 0
horizon = 1
industry_level = 1
industry_name = 'kh'
universe = 'enliven_fst'
begin_date = '2021-01-01'
end_date = '2021-08-10'
```

```python
#### 回调函数
def callback_save(gen, rootid, best_programs, custom_params):
    print(best_programs)
```

```python
### 若设置callback_save回调 保存每一代最优因子回调函数。 如果不设置默认为None，通过阻塞方式即run函数返回所有进化代数最优因子。
gentic = EntropyAPI.FuturesGeneticist(offset=0,
                                      horizon=1,
                                      factor_columns=factor_columns,
                                      universe=universe,
                                      industry_name=industry_name,
                                      industry_level=industry_level,
                                      operators=operators_sets,
                                      callback_save=None)
```

```python
evolution_sets = gentic.run(begin_date=begin_date,
                   end_date=end_date,
                   configure=configure)
```

```python
for program in evolution_sets.values():
    print("eval:{0},metrics:{1}\n".format(program.transform(), program._retain_data))
```
