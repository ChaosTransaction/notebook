获取交易日历，本例子使用沪深交易时间,还包括有商品期货和港股交易时间

```python
from ultron.tradingday import *
```

```python
begin_date ='2020-01-01'
end_date = '2022-03-10'
```

### 交易时间区间

##### 连续交易日

```python
dates = makeSchedule(begin_date,end_date, '1b', 'china.sse',
                         BizDayConventions.Preceding)
dates
```

##### 每隔四天交易日

```python
dates = makeSchedule(begin_date,end_date, '4b', 'china.sse',
                         BizDayConventions.Preceding)
dates
```

### 交易日推算

##### 向后推算最近的两个交易日

```python
advanceDateByCalendar('china.sse', begin_date, '2b')
```

##### 向前推算最近的三个交易日

```python
advanceDateByCalendar('china.sse', begin_date, '-3b')
```

### 判断是否为交易日

```python
isBizDay('china.sse', begin_date)
```
